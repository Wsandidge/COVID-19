```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
conf = pd.read_csv('US_Confirmed.csv')
dead = pd.read_csv('US_Deaths.csv')
```

```python
x=conf['Key']
y=conf['Confirmed Cases']

x1=dead['Key']
y1=dead['Deaths']
```

```python
p = np.poly1d(np.polyfit(x, y, 6))
t=np.arange(0,90,.01)
plt.plot(x, y, 'o', t, p(t), '-')

plt.title('US COVID-19 Confirmed Cases')
plt.grid()
plt.savefig('3-18_Confirmed_1-month.jpeg', dpi=600)
```

```python
p1 = np.poly1d(np.polyfit(x1, y1, 6))

plt.plot(x1, y1, 'o', t, p1(t), '-')

plt.title('US COVID-19 Deaths')
plt.grid()
plt.savefig('3-18_Deaths_1-month.jpeg', dpi=600)
```
